```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
embedding_2d = np.genfromtxt('embedding_2d.csv', delimiter = ',')
assigned_label = np.genfromtxt('assigned_label_umap3d.csv', delimiter = ',')
```

```python
colors = cm.spring(np.linspace(0, 1, 4))
```

```python
fig, axs = plt.subplots(1, 1, dpi = 300)
fig.set_size_inches(6, 5)

axs.set_title("4 disconnected manifolds")
axs.set_xticks([])
axs.set_yticks([])

axs.scatter(embedding_2d[np.where(assigned_label == 2)[0], 0], embedding_2d[np.where(assigned_label == 2)[0], 1], s = 5, color = colors[0], label = 'wild type')

for k in [4, 8, 21, 11, 16, 10, 22]:

    axs.scatter(embedding_2d[np.where(assigned_label == k)[0], 0], embedding_2d[np.where(assigned_label == k)[0], 1], s = 5, color = colors[0])

axs.scatter(embedding_2d[np.where(assigned_label == 9)[0], 0], embedding_2d[np.where(assigned_label == 9)[0], 1], s = 5, color = colors[2], label = '16187_16189TAC>CCT')    
    
for k in [24, 7, 26, 14, 3, 5, 17, 13, 12, 19, 23, 20, 18, 6, 15, 0, 27, 1, 25]:

    axs.scatter(embedding_2d[np.where(assigned_label == k)[0], 0], embedding_2d[np.where(assigned_label == k)[0], 1], s = 5, color = colors[2])    

for k in [29]:

    axs.scatter(embedding_2d[np.where(assigned_label == k)[0], 0], embedding_2d[np.where(assigned_label == k)[0], 1], s = 5, color = colors[1], label = '2788C>A') 

axs.scatter(embedding_2d[np.where(assigned_label == 28)[0], 0], embedding_2d[np.where(assigned_label == 28)[0], 1], s = 5, color = colors[3], label = '2784A>T, 2788C>A')    
    
for k in [30]:

    axs.scatter(embedding_2d[np.where(assigned_label == k)[0], 0], embedding_2d[np.where(assigned_label == k)[0], 1], s = 5, color = colors[3]) 

axs.legend(loc ="lower right", fontsize = "10")
    
plt.show()
```
