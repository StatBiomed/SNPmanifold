```python
import seaborn as sns

reference = np.empty(assigned_label.shape[0])
reference[:4339] = 0
reference[4339:] = 1
reference = reference.astype(int)
predict = assigned_label.astype(int)
    
matrix = np.zeros((2, cluster_no))
ref_sum = np.zeros(cluster_no)

for q in range(assigned_label.shape[0]):

        matrix[reference[q], predict[q]] += 1
        ref_sum[predict[q]] += 1

prop = matrix / np.outer(np.ones(2), ref_sum)

fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15, 10)
ax = sns.heatmap(prop[:, np.flip(np.argsort(prop[0, :]))].T, xticklabels = ["HSPCs", "PBMCs"], yticklabels = np.flip(np.argsort(prop[0, :])), vmin = 0, vmax = 1)
ax.set_ylabel("Cluster", fontsize = 35)
ax.set_title("Confusion", fontsize = 40)
plt.gcf().set_dpi(300)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 35)
plt.show()
```
